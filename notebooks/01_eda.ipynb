# NFL Big Data Bowl 2026 - EDA
Exploratory Data Analysis of tracking data.

```python
import sys
sys.path.append('../src')
from data_loader import DataLoader
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize loader
loader = DataLoader(data_dir='/home/tanmay/Desktop/NFL')
df = loader.load_week_data(1)
print(f"Loaded {df.height} rows.")
df.head()
```

## 1. Standardization Check
Verify that plays going 'left' have been flipped correctly.

```python
df = loader.standardize_tracking_directions(df)
df.select(['game_id', 'play_id', 'play_direction', 'x', 'y', 'std_x', 'std_y']).head()
```

## 2. Visualization
Plot a single play.

```python
play_df = df.filter((pl.col("game_id") == df["game_id"][0]) & (pl.col("play_id") == df["play_id"][0]))
play_df_pd = play_df.to_pandas()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=play_df_pd, x='std_x', y='std_y', hue='club')
plt.title(f"Play {play_df_pd['play_id'].iloc[0]}")
plt.show()
```
